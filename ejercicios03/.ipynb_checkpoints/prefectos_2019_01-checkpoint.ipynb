```python
# generar el proceso para prefectos 
# permitir buscar por provincias
import pandas as pd
import ipywidgets as widgets
```

```python
datos_prefectos = pd.read_spss('data/resultadosdignidadprefecto2019.sav')
datos_prefectos
```

```python
datos_candidatos = pd.read_spss('data/candidatos2019.sav')
datos_candidatos.shape
```

```python
datos_candidatos.head()
```

```python
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
datos_candidatos_2.shape
```

```python
datos_prefectos = pd.merge(datos_prefectos, datos_candidatos_2, on="CANDIDATO_CODIGO")
datos_prefectos
```

```python
organizaciones_polticas = pd.read_spss("data/organizacionespolticas2019.sav")
organizaciones_polticas.keys()
```

```python
organizaciones_polticas_2 = organizaciones_polticas[['OP_CODIGO', 'OP_NOMBRE']]
organizaciones_polticas_2
```

```python
datos_prefectos = pd.merge(datos_prefectos, organizaciones_polticas_2, on="OP_CODIGO")
datos_prefectos
```

```python
datos_loja = datos_prefectos[datos_prefectos['PROVINCIA_NOMBRE']=="LOJA"]
datos_loja
```

```python
datos_loja.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values().plot(kind="bar")
```
